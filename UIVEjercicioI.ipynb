# Distribucion Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
num_datos = 500

# Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')# Histograma de datos
plt.title('Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.grid(True)
plt.show()
```

```python
# Crear un rango de valores para crear una densidad
x = np.linspace(media-3*desv_std, media+3*desv_std, num_datos)

# Calcular la PDF de la distribucion normal
pdf = 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2) )
```

```python
print("Imprimir los Primeros 50 Valores de x", x[0:50], " y los Primeros 50 Valores de pdf", pdf[0:50])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.grid(True)
plt.show()
```

## Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:
## La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
## Cuantos estudiantes pesan más de 68 y menos de 72 kg.

```python
import scipy.stats as stats

media = 70
desviacion_estandar = 4
n_estudiantes = 500

p_menor_72 = stats.norm.cdf(72, media, desviacion_estandar)
p_menor_68 = stats.norm.cdf(68, media, desviacion_estandar)
probabilidad = p_menor_72 - p_menor_68

cantidad_estudiantes = probabilidad * n_estudiantes

print(f"Probabilidad de estudiantes que pesan entre 68 y 72 kg: {probabilidad:.4f}")
print(f"Cantidad de estudiantes que pesan entre 68 y 72 kg: {cantidad_estudiantes:.0f}")


pesos = np.random.normal(media, desviacion_estandar, n_estudiantes)

plt.hist(pesos, bins=20, edgecolor='black', alpha=0.75, density=True)

plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia relativa')
plt.title('Distribución de pesos de los estudiantes')

plt.show()
```
