# 1.Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from enum import auto
auto = pd.read_csv('/content/autos2.csv')
```

```python
print(auto.head())
```

```python
minimo = auto['price'].min()
print(minimo)
```

```python
maximo = auto['price'].max()
print(maximo)
```

```python
moda = auto['price'].mode()[0]
print(moda)
```

```python
media = auto['price'].mean()
print(media)
```

```python
mediana = auto['price'].median()
print(mediana)
```

```python
desviacion_estandar_precio = auto['price'].std()
print (desviacion_estandar_precio)
```

```python
auto.price.describe()
```

# Histograma de precios de los automóviles

```python
plt.hist(auto['price'], bins=10, edgecolor='black')
plt.title('Histograma de Precios de Automóviles')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

# Conclusion
El gráfico muestra la distribución del precio. Esto puede saber si es más bajo o más frecuente en la parte superior del rango de precios.

# Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Tipo de Vehículo')
plt.xlabel('Tipo de Vehículo')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Explica cómo cambian los precios dependiendo del tipo de cuerpo del automóvil, qué tipo de vehículos es más costoso y el tipo de vehículos que parecen baratos.

# Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Tipo de Tracción')
plt.xlabel('Tipo de Tracción')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Esto nos ayuda a seguir la diferencia de precio según el tipo de automóvil, lo que puede confirmar que los sistemas de tracción especificados son más caros.

# Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y la ubicación del motor (engine-location)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Ubicación del Motor')
plt.xlabel('Ubicación del Motor')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Proporcione cómo el lugar del motor afecta los precios, proporcionando la idea de la configuración de diseño más costosa.
