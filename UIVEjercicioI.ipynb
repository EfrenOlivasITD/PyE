# Ejercicio 1
Una revision de los pesos de  500 estudiantes del itd, evidencio que la media es de 70kg y la desviacione estandar es de 4kg. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:


*   Generar aleatoriamente las edades de los 500 estudiantes con una media de 70kg y una desviacion estandar de 4kg.
*   Realizar un histograma de las edades


*   Graficar la funcion de densidad de probabilidad (PDF)





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribucion normal
media = 70 # Media
desv_std = 4 # Desviacion estandar
num_datos = 500

# Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] #Mostrar los primeros 50
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')# Histograma de datos
plt.title('Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.grid(True)
plt.show()
```

```python
# Crear un rango de valores para crear una densidad
x = np.linspace(media-3*desv_std, media+3*desv_std, num_datos)

# Calcular la PDF de la distribucion normal
pdf = 1/(desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2) )
```

```python
print("Imprimir los Primeros 50 Valores de x", x[0:50], " y los Primeros 50 Valores de pdf", pdf[0:50])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='normal:(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Funcion de Densidad de Probabilidad (PDF) - Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2, )

plt.title('Distribucion Normal')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

# Parametros de la distribucion normal
media = 70 #Media
desv_std = 4 #Desviacion estandar
total_estudiantes = 500

# Variable para calculo de la probabilidad de que un estudainte pese menos de 75kg
peso_limite = 79

# Primero necesitamos calcular la probabilidad que un estudainte pese menos
# de 75kg utilizando la funcion de distribucion acumulativa (CDF) de la distribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
# Calculo del numero de estudiantesque pesan menos de 75kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un Estudiante Pese Menos de 75kg: ", prob_menos_75kg)
print("Numero de Estudiantes que Pesen Menos de 75kg: ", estudiantes_menos_75kg)

# Graficar la distribucion normal
x = np.linspace(media-4*desv_std, media+4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt. fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$,  x=500')
plt.xlabel('Peso en Kg de los Estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.show()
```
