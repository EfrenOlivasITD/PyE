# 1.Cargar el Dataset de Autos y realizar las Estadsticas descriptivas bsicas y medidas de tendencia central

```python
minimo = auto['price'].min()
print(minimo)
```

# Histograma de precios de los automviles

```python
plt.hist(auto['price'], bins=10, edgecolor='black')
plt.title('Histograma de Precios de Automviles')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

# Conclusion
El grfico muestra la distribucin del precio. Esto puede saber si es ms bajo o ms frecuente en la parte superior del rango de precios.

# Diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo de vehculo (body style)

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='body-style', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Tipo de Vehculo')
plt.xlabel('Tipo de Vehculo')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Explica cmo cambian los precios dependiendo del tipo de cuerpo del automvil, qu tipo de vehculos es ms costoso y el tipo de vehculos que parecen baratos.

# Diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y tipo traccin ruedas (drive wheels)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Tipo de Traccin')
plt.xlabel('Tipo de Traccin')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Esto nos ayuda a seguir la diferencia de precio segn el tipo de automvil, lo que puede confirmar que los sistemas de traccin especificados son ms caros.

# Diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automvil (price) y la ubicacin del motor (engine-location)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=auto)
plt.title('Diagrama de Caja y Bigotes de Precios por Ubicacin del Motor')
plt.xlabel('Ubicacin del Motor')
plt.ylabel('Precio')
plt.show()
```

# Conclusion
Proporcione cmo el lugar del motor afecta los precios, proporcionando la idea de la configuracin de diseo ms costosa.
